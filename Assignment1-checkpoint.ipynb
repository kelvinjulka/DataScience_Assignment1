```python
for col in df.columns.values:
    if df[col].isnull().sum() == 0:
        continue
    if col == 'Climate':
        guess_values = df.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
    else:
        guess_values = df.groupby('Region')[col].median()
    for region in df['Region'].unique():
        df[col].loc[(df[col].isnull())&(df['Region']==region)] = guess_values[region]
```

We are replacing null values from dataset with median of respective region.

```python
print(df.isnull().sum())
```

We have just confirmed there is no Null value.

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20,20))
plt.subplots_adjust(hspace=0.4)

corr_to_gdp = pd.Series()
for col in df.columns.values[2:]:
    if ((col!='GDP ($ per capita)')&(col!='Climate')):
        corr_to_gdp[col] = df['GDP ($ per capita)'].corr(df[col])
abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)
corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]

for i in range(3):
    for j in range(3):
        sns.regplot(x=corr_to_gdp.index.values[i*3+j], y='GDP ($ per capita)', data=df,
                   ax=axes[i,j], fit_reg=False, marker='.')
        title = 'correlation='+str(corr_to_gdp[i*3+j])
        axes[i,j].set_title(title)
axes[1,2].set_xlim(0,102)
plt.show()
```

Representation of graph ,fators affecting GDP:
1.Relation between GDP and Phones are directly propotional to each other.
2.Relation between GDP and Birthrate is inversly propotional to each other.
3.Relation between GDP and Infant Morality rate is inversly propotional to each other.
4.Relation between GDP and Aggriculture is inversly propotional to each other.
5.Relation between GDP and Service  are directly propotional to each other.
6.Relation between GDP and net migration  is inversly  propotional to each other.
7..Relation between GDP and Crops  is inversly  propotional to each other.
8.Relation between GDP and Death rate is inversly  propotional to each other.
